```python
import pandas as pd
import random
import numpy as np
```

```python
days = 365 * 5 + 30
dates = pd.date_range('20171201', periods=days)
```

```python
dfopen = pd.DataFrame(np.random.rand(days)*10 + 1533, index=dates, columns=['Open'] )
dfopen = dfopen.round(2)
```

```python
dfhigh = dfopen.applymap(lambda x: x + np.random.rand()*10)
dfhigh.columns = ['High']
dfhigh = dfhigh.round(2)
```

```python
dflow = dfopen.applymap(lambda x: x - np.random.rand()*10)
dflow.columns = ['Low']
dflow = dflow.round(2)
```

```python
dfclose = dflow.applymap(lambda x: x + np.random.rand()*2)
dfclose = dfclose.round(2)
dfclose.columns = ['Close']
```

```python
dfchange = pd.DataFrame(columns=['Change'])
dfchange['Change'] = dfclose['Close'] - dfopen['Open']
dfchange = dfchange.round(2)
```

```python
dfvolume = pd.DataFrame(random.sample(range(0, 52676), days), index=dates, columns=['Volume'] )
```

```python
dfx = dfopen.join(dfhigh)
dfx = dfx.join(dflow)
dfx = dfx.join(dfclose)
dfx = dfx.join(dfchange)
dfx = dfx.join(dfvolume)
dfx.index.name = "Date"
```
